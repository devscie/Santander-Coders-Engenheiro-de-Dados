```python
import csv
```

```python
with open('brasil_covid.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv, delimiter=',')
    for linha in leitor:
        print(linha)
```

```python
with open('brasil_covid.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv)
    next(leitor) #pular o cabeçalho
    for linha in leitor:
        if float(linha[3]) > 1:
            print(linha)
```

```python
# Leitura sem biblioteca
with open('brasil_covid.csv', 'r') as csv_file:
    linhas = csv_file.read()
    linhas = linhas.split('\n')
    for linha in linhas:
        linha = linha.split(',')
        print(linha)
```

```python
import csv
with open('brasil_covid.csv', newline='') as csvfile:
    leitor = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for linha in leitor:
        print(', '.join(linha))
```

```python
import csv

with open('users.csv', 'w', newline='') as arquivo_users:
    escritor = csv.writer(arquivo_users, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    escritor.writerow(['nome', 'sobrenome', 'email', 'genero'])
    escritor.writerow(['João', 'Keppler', 'jkeppler@email.com.br', 'Masculino'])
```

```python
#Cadastro

import csv

header = ['nome', 'sobrenome']
dados = []
opt = input('O que deseja fazer?\n1 - Cadastrar\n0 - Sair\n')
while opt != '0':
    nome = input('Qual seu nome?')
    sobrenome = input('Qual seu sobrenome?')
    dados.append([nome, sobrenome])
    opt = input('O que deseja fazer?\n1 - Cadastrar\n0 - Sair\n') 
    
print(dados)

with open('users.csv', 'w', newline='') as arquivo_csv:
    writer = csv.writer(arquivo_csv)
    writer.writerow(header)
    writer.writerows(dados)
    
with open('users.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        print(row)
```
